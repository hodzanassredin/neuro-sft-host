```python
import os
import datasets
import re
```

```python
dataset_path = "/app/datasets/oberon/docs/bb"
dataset_name = 'hodza/BlackBox-2.0-a1'
```

```python
def load_texts():
    file_names = [] 
    for subdir, dirs, files in os.walk(dataset_path):
        if "/Rsrc/" in subdir:
            continue
        for file in files:
            file_names.append(os.path.join(subdir, file))
    texts = []
    for f in file_names:
        with open(f, 'r', encoding='utf-8') as file:
            text = file.read()
            text = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f\x7f-\xff]', '', text)
            texts.append(text)
    return file_names,texts
names, texts = load_texts()

```

```python
names = [n[len(dataset_path)+1:] for n in names]
text_dataset = datasets.Dataset.from_dict({"texts":texts,"names" : names})
```

```python
text_dataset
```

```python

text_dataset.push_to_hub(dataset_name)
```
