```python
import os

def rename_txt_to_md(directory):
    # Loop through all files in the specified directory
    for filename in os.listdir(directory):
        # Check if the file ends with .txt
        if filename.endswith(".txt"):
            # Create the new filename by replacing .txt with .md
            new_filename = filename.replace(".txt", ".md")
            
            # Get the full paths for the old and new filenames
            old_file = os.path.join(directory, filename)
            new_file = os.path.join(directory, new_filename)
            
            # Rename the file
            os.rename(old_file, new_file)
            print(f"Renamed: {filename} -> {new_filename}")

# Specify the directory containing the .txt files
output_folder = 'txt_files3'

# Call the function to rename the files
rename_txt_to_md(output_folder)
```
