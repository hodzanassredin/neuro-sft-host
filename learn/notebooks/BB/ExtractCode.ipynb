```python
import re

def load_file(fname):
    with open(fname, 'r', encoding='utf-8') as file:
        return file.readlines()

def is_empty_line(line):
    return not line.strip()

def highlight_code(lines):
    highlighted_lines = []
    
    # Define regex patterns for different constructs
    module_pattern = re.compile(r'^\s*MODULE\s+\w+\s*;')
    proc_decl_pattern = re.compile(r'^\s*PROCEDURE\s+.*')
    forward_decl_pattern = re.compile(r'^\s*PROCEDURE\s+\^\s+.*')
    comment_pattern = re.compile(r'^\s*\(\*(.*?)\*\)\s*$')
    
    for line in lines:
        stripped_line = line.strip()
        
        if is_empty_line(line):
            highlighted_lines.append("Empty")
        elif module_pattern.match(stripped_line):
            highlighted_lines.append("Module")
        elif proc_decl_pattern.match(stripped_line):
            highlighted_lines.append("Procedure")
        elif forward_decl_pattern.match(stripped_line):
            highlighted_lines.append("Forward Declaration")
        elif comment_pattern.match(stripped_line):
            highlighted_lines.append("Comment")
        else:
            highlighted_lines.append("Code")
    
    return highlighted_lines

def main(fname):
    lines = load_file(fname)
    highlighted_lines = highlight_code(lines)
    
    for line, label in zip(lines, highlighted_lines):
        print(f"{label}: {line.strip()}")

# Example usage
main(fname)
```
