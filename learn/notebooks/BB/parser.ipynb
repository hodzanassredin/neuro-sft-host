```python
import ply.lex as lex
import ply.yacc as yacc

# Токены
tokens = (
    'MODULE', 'IMPORT', 'CONST', 'TYPE', 'VAR', 'PROCEDURE', 'BEGIN', 'END', 'CLOSE',
    'NEW', 'ABSTRACT', 'EMPTY', 'EXTENSIBLE', 'ARRAY', 'OF', 'RECORD', 'POINTER', 'TO',
    'IF', 'THEN', 'ELSIF', 'ELSE', 'CASE', 'WHILE', 'DO', 'REPEAT', 'UNTIL', 'FOR', 'BY',
    'LOOP', 'WITH', 'EXIT', 'RETURN', 'IN', 'OUT', 'NIL', 'DIV', 'MOD', 'OR', 'AND',
    'IDENT', 'NUMBER', 'CHARACTER', 'STRING', 'SEMICOLON', 'COMMA', 'COLON', 'EQUALS',
    'LPAREN', 'RPAREN', 'LBRACKET', 'RBRACKET', 'DOT', 'PLUS', 'MINUS', 'TIMES', 'DIVIDE',
    'LESS', 'LESSEQUAL', 'GREATER', 'GREATEREQUAL', 'NOTEQUAL', 'ASSIGN', 'ARROW', 'HAT',
    'DOLLAR', 'LBRACE', 'RBRACE', 'BAR'
)

# Регулярные выражения для токенов
t_MODULE = r'MODULE'
t_IMPORT = r'IMPORT'
t_CONST = r'CONST'
t_TYPE = r'TYPE'
t_VAR = r'VAR'
t_PROCEDURE = r'PROCEDURE'
t_BEGIN = r'BEGIN'
t_END = r'END'
t_CLOSE = r'CLOSE'
t_NEW = r'NEW'
t_ABSTRACT = r'ABSTRACT'
t_EMPTY = r'EMPTY'
t_EXTENSIBLE = r'EXTENSIBLE'
t_ARRAY = r'ARRAY'
t_OF = r'OF'
t_RECORD = r'RECORD'
t_POINTER = r'POINTER'
t_TO = r'TO'
t_IF = r'IF'
t_THEN = r'THEN'
t_ELSIF = r'ELSIF'
t_ELSE = r'ELSE'
t_CASE = r'CASE'
t_WHILE = r'WHILE'
t_DO = r'DO'
t_REPEAT = r'REPEAT'
t_UNTIL = r'UNTIL'
t_FOR = r'FOR'
t_BY = r'BY'
t_LOOP = r'LOOP'
t_WITH = r'WITH'
t_EXIT = r'EXIT'
t_RETURN = r'RETURN'
t_IN = r'IN'
t_OUT = r'OUT'
t_NIL = r'NIL'
t_DIV = r'DIV'
t_MOD = r'MOD'
t_OR = r'OR'
t_AND = r'AND'
t_SEMICOLON = r';'
t_COMMA = r','
t_COLON = r':'
t_EQUALS = r'='
t_LPAREN = r'\('
t_RPAREN = r'\)'
t_LBRACKET = r'\['
t_RBRACKET = r'\]'
t_DOT = r'\.'
t_PLUS = r'\+'
t_MINUS = r'-'
t_TIMES = r'\*'
t_DIVIDE = r'/'
t_LESS = r'<'
t_LESSEQUAL = r'<='
t_GREATER = r'>'
t_GREATEREQUAL = r'>='
t_NOTEQUAL = r'\#'
t_ASSIGN = r':='
t_ARROW = r'\^'
t_HAT = r'\^'
t_DOLLAR = r'\$'
t_LBRACE = r'\{'
t_RBRACE = r'\}'
t_BAR = r'\|'

# Идентификаторы и литералы
def t_IDENT(t):
    r'[a-zA-Z_][a-zA-Z_0-9]*'
    return t

def t_NUMBER(t):
    r'\d+'
    t.value = int(t.value)
    return t

def t_CHARACTER(t):
    r"'.'"
    return t

def t_STRING(t):
    r'"[^"]*"'
    return t

# Пропускать пробелы и табуляции
t_ignore = ' \t'

# Обработка ошибок
def t_error(t):
    print(f"Illegal character '{t.value[0]}'")
    t.lexer.skip(1)

# Создание лексического анализатора
lexer = lex.lex()

# Правила грамматики
def p_module(p):
    '''module : MODULE ident SEMICOLON import_list decl_seq begin_stmt_seq close_stmt_seq END ident DOT'''
    pass

def p_import_list(p):
    '''import_list : IMPORT import_ident_list SEMICOLON
                  | empty'''
    pass

def p_import_ident_list(p):
    '''import_ident_list : ident
                         | ident COLONEQUALS ident
                         | import_ident_list COMMA ident
                         | import_ident_list COMMA ident COLONEQUALS ident'''
    pass

def p_decl_seq(p):
    '''decl_seq : const_decl_seq type_decl_seq var_decl_seq proc_decl_seq forward_decl_seq
                | empty'''
    pass

def p_const_decl_seq(p):
    '''const_decl_seq : CONST const_decl_list SEMICOLON
                      | empty'''
    pass

def p_const_decl_list(p):
    '''const_decl_list : const_decl
                       | const_decl_list const_decl'''
    pass

def p_const_decl(p):
    '''const_decl : ident_def EQUALS const_expr'''
    pass

def p_type_decl_seq(p):
    '''type_decl_seq : TYPE type_decl_list SEMICOLON
                     | empty'''
    pass

def p_type_decl_list(p):
    '''type_decl_list : type_decl
                      | type_decl_list type_decl'''
    pass

def p_type_decl(p):
    '''type_decl : ident_def EQUALS type'''
    pass

def p_var_decl_seq(p):
    '''var_decl_seq : VAR var_decl_list SEMICOLON
                    | empty'''
    pass

def p_var_decl_list(p):
    '''var_decl_list : var_decl
                     | var_decl_list var_decl'''
    pass

def p_var_decl(p):
    '''var_decl : ident_list COLON type'''
    pass

def p_proc_decl_seq(p):
    '''proc_decl_seq : proc_decl
                     | proc_decl_seq proc_decl'''
    pass

def p_proc_decl(p):
    '''proc_decl : PROCEDURE receiver ident_def formal_pars meth_attributes SEMICOLON decl_seq begin_stmt_seq END ident'''
    pass

def p_meth_attributes(p):
    '''meth_attributes : COMMA NEW
                       | COMMA ABSTRACT
                       | COMMA EMPTY
                       | COMMA EXTENSIBLE
                       | empty'''
    pass

def p_forward_decl_seq(p):
    '''forward_decl_seq : forward_decl
                        | forward_decl_seq forward_decl'''
    pass

def p_forward_decl(p):
    '''forward_decl : PROCEDURE ARROW receiver ident_def formal_pars meth_attributes'''
    pass

def p_formal_pars(p):
    '''formal_pars : LPAREN fp_section_list RPAREN COLON type
                   | LPAREN fp_section_list RPAREN
                   | empty'''
    pass

def p_fp_section_list(p):
    '''fp_section_list : fp_section
                       | fp_section_list SEMICOLON fp_section'''
    pass

def p_fp_section(p):
    '''fp_section : VAR ident_list COLON type
                  | IN ident_list COLON type
                  | OUT ident_list COLON type
                  | ident_list COLON type'''
    pass

def p_receiver(p):
    '''receiver : LPAREN VAR ident COLON ident RPAREN
                | LPAREN IN ident COLON ident RPAREN
                | empty'''
    pass

def p_type(p):
    '''type : qualident
            | ARRAY const_expr_list OF type
            | RECORD field_list_seq END
            | POINTER TO type
            | PROCEDURE formal_pars'''
    pass

def p_field_list_seq(p):
    '''field_list_seq : field_list
                       | field_list_seq SEMICOLON field_list'''
    pass

def p_field_list(p):
    '''field_list : ident_list COLON type
                  | empty'''
    pass

def p_const_expr_list(p):
    '''const_expr_list : const_expr
                       | const_expr_list COMMA const_expr'''
    pass

def p_begin_stmt_seq(p):
    '''begin_stmt_seq : BEGIN statement_seq
                      | empty'''
    pass

def p_close_stmt_seq(p):
    '''close_stmt_seq : CLOSE statement_seq
                      | empty'''
    pass

def p_statement_seq(p):
    '''statement_seq : statement
                     | statement_seq SEMICOLON statement'''
    pass

def p_statement(p):
    '''statement : designator ASSIGN expr
                 | designator LPAREN expr_list RPAREN
                 | IF expr THEN statement_seq elsif_stmt_seq else_stmt_seq END
                 | CASE expr OF case_list else_stmt_seq END
                 | WHILE expr DO statement_seq END
                 | REPEAT statement_seq UNTIL expr
                 | FOR ident ASSIGN expr TO expr by_const_expr DO statement_seq END
                 | LOOP statement_seq END
                 | WITH guard_stmt_seq else_stmt_seq END
                 | EXIT
                 | RETURN expr'''
    pass

def p_elsif_stmt_seq(p):
    '''elsif_stmt_seq : ELSIF expr THEN statement_seq
                      | elsif_stmt_seq ELSIF expr THEN statement_seq'''
    pass

def p_else_stmt_seq(p):
    '''else_stmt_seq : ELSE statement_seq
                     | empty'''
    pass

def p_case_list(p):
    '''case_list : case
                 | case_list BAR case'''
    pass

def p_case(p):
    '''case : case_labels COLON statement_seq'''
    pass

def p_case_labels(p):
    '''case_labels : const_expr
                  | const_expr DOTDOT const_expr'''
    pass

def p_guard_stmt_seq(p):
    '''guard_stmt_seq : guard DO statement_seq
                       | guard_stmt_seq BAR guard DO statement_seq'''
    pass

def p_guard(p):
    '''guard : qualident COLON qualident'''
    pass

def p_const_expr(p):
    '''const_expr : expr'''
    pass

def p_expr(p):
    '''expr : simple_expr
            | simple_expr relation simple_expr'''
    pass

def p_simple_expr(p):
    '''simple_expr : term
                   | simple_expr add_op term'''
    pass

def p_term(p):
    '''term : factor
            | term mul_op factor'''
    pass

def p_factor(p):
    '''factor : designator
              | NUMBER
              | CHARACTER
              | STRING
              | NIL
              | set
              | LPAREN expr RPAREN
              | NOT factor'''
    pass

def p_set(p):
    '''set : LBRACE element_list RBRACE'''
    pass

def p_element_list(p):
    '''element_list : element
                    | element_list COMMA element'''
    pass

def p_element(p):
    '''element : expr
               | expr DOTDOT expr'''
    pass

def p_relation(p):
    '''relation : EQUALS
                | NOTEQUAL
                | LESS
                | LESSEQUAL
                | GREATER
                | GREATEREQUAL
                | IN
                | IS'''
    pass

def p_add_op(p):
    '''add_op : PLUS
             | MINUS
             | OR'''
    pass

def p_mul_op(p):
    '''mul_op : TIMES
              | DIVIDE
              | DIV
              | MOD
              | AND'''
    pass

def p_designator(p):
    '''designator : qualident
                  | designator DOT ident
                  | designator LBRACKET expr_list RBRACKET
                  | designator ARROW
                  | designator LPAREN qualident RPAREN
                  | designator LPAREN expr_list RPAREN
                  | designator DOLLAR'''
    pass

def p_expr_list(p):
    '''expr_list : expr
                 | expr_list COMMA expr'''
    pass

def p_ident_list(p):
    '''ident_list : ident_def
                  | ident_list COMMA ident_def'''
    pass

def p_qualident(p):
    '''qualident : ident
                 | ident DOT ident'''
    pass

def p_ident_def(p):
    '''ident_def : ident
                 | ident TIMES
                 | ident MINUS'''
    pass

def p_ident(p):
    '''ident : IDENT'''
    pass

def p_by_const_expr(p):
    '''by_const_expr : BY const_expr
                     | empty'''
    pass

def p_empty(p):
    '''empty :'''
    pass

# Обработка ошибок синтаксического анализа
def p_error(p):
    if p:
        print(f"Syntax error at '{p.value}'")
    else:
        print("Syntax error at EOF")

# Создание парсера
parser = yacc.yacc()

# Функция для проверки строки
def is_valid_code(code):
    try:
        parser.parse(code, lexer=lexer)
        return True
    except:
        return False

# Пример использования
code = """
MODULE Example;
IMPORT Math;
CONST Pi = 3.14;
TYPE Point = RECORD x, y: INTEGER END;
VAR p: Point;
PROCEDURE Init(VAR p: Point);
BEGIN
    p.x := 0;
    p.y := 0;
END Init;
BEGIN
    Init(p);
END Example.
"""

print(is_valid_code(code))  # Должно вернуть True, если строка является корректным кодом
```
